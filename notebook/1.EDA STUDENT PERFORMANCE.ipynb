```python
## define numerical and categorical columns
numerical_features=[feature for feature in df.columns if df[feature].dtype!='O']
categorical_features=[feature for feature in df.columns if df[feature].dtype=='O']
```

```python
print('We have {} numerical features : {}'.format(len(numerical_features),numerical_features))
print('We have {} categorical features : {}'.format(len(categorical_features),categorical_features))
```

```python
## Adding columns for "Total Score" and "Average"

df['Total_score']=df['math_score']+df['reading_score']+df['writing_score']
df['Average']=df['Total_score']/3
df.head()
```

```python
## calculate the number of students got full marks

maths_full=df[df['math_score']==100]['Average'].count()
reading_full=df[df['reading_score']==100]['Average'].count()
writing_full=df[df['writing_score']==100]['Average'].count()

print("Total number of students who score 100 marks in maths : ",maths_full)
print("Total number of students who score 100 marks in reading : ",reading_full)
print("Total number of students who score 100 marks in writing : ",writing_full)
```

```python
## number of students who score less than 20

maths_less=df[df['math_score']<=20]['Average'].count()
reading_less=df[df['reading_score']<=20]['Average'].count()
writing_less=df[df['writing_score']<=20]['Average'].count()

print("Total number of students who score less than 20 in maths : ",maths_less)
print("Total number of students who score less than 20 in reading : ",reading_less)
print("Total number of students who score less than 20 in writing : ",writing_less)
```

```python
numeric_df = df.select_dtypes(include='number')

# Perform the groupby and mean operations
gender_group = df.groupby('gender')[numeric_df.columns].mean()

# Display the result
print(gender_group)
```

```python
plt.figure(figsize=(10, 8))

X = ['Total Average','Math Average']


female_scores = [gender_group['Average'][0], gender_group['math_score'][0]]
male_scores = [gender_group['Average'][1], gender_group['math_score'][1]]

X_axis = np.arange(len(X))
  
plt.bar(X_axis - 0.2, male_scores, 0.4, label = 'Male')
plt.bar(X_axis + 0.2, female_scores, 0.4, label = 'Female')
  
plt.xticks(X_axis, X)
plt.ylabel("Marks")
plt.title("Total average v/s Math average marks of both the genders", fontweight='bold')
plt.legend()
plt.show()
```
